```python
import matplotlib.pyplot as plt
```

```python
years = [1993, 1994, 1995, 1996, 1997]
pathways_count = [33, 42, 59, 68, 39]
linear_count = [249, 291,  338, 377, 345]

plt.plot(years, pathways_count, label='Pathways')
plt.plot(years, linear_count, label='Linear')
plt.title('Homeless Count by Year', loc='left')
plt.xlabel('Year')
plt.legend()
plt.xticks(years)
plt.show()
```
