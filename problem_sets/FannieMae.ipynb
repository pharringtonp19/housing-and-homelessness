### **Import Libraries**

```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
```

### **Question #1**

```python
def f1(r):
  return (1+r)**(1/12) - 1

def f2(r):
  return r/12

rs = jnp.linspace(0.01, 0.10, 100)
ys1 = f1(rs)
ys2 = f2(rs)
```

```python
plt.plot(rs, ys1, label='Approximate')
plt.plot(rs, ys2, label='True')
plt.xlabel('Annual Interest Rate')
plt.title('Monthly Interest Rate', loc='left')
plt.legend()
plt.show()
```

### **Question #2**

```python
def MonthlyPayment(n, r, p):
  numerator = p
  denominator = jnp.sum(jnp.array([1/(1+r/12)**i for i in range(1, n+1)]))
  return numerator/denominator
```

```python
n = 12*30
r = 0.05
p = 100_000

rs = jnp.linspace(0.01, 0.1, 100)
ys = jax.vmap(MonthlyPayment, in_axes=(None, 0, None))(n, rs, p)
```

```python
plt.plot(rs, ys)
plt.xlabel('Annual Interest Rate')
plt.title('Monthly Payment', loc='left')
plt.show()
```
